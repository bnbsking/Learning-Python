```python
from random import shuffle

v = [ ['O']*8, ['O']*8, ['O']*8, ['O']*8 ]
chess = ['b將', 'b士', 'b士', 'b象', 'b象', 'b車', 'b車', 'b馬', 'b馬', 'b包', 'b包', 'b卒',\
    'b卒', 'b卒', 'b卒', 'b卒', 'r帥', 'r仕', 'r仕', 'r相', 'r相', 'r俥', 'r俥', 'r傌', 'r傌',\
    'r炮', 'r炮', 'r兵', 'r兵', 'r兵', 'r兵', 'r兵']
shuffle(chess)

mine, flag = "", False
```

```python
def printv():
    for row in v:
        print(row)
    print(mine, "\n")

def check(row, col, motion):
    try:
        row, col, motion = int(row), int(col), int(motion)# int transfer and transferable check
        if v[row][col] != "_":                            # row col range ckeck + non-null check 
            if motion==0 and v[row][col]=="O":            # back check
                return row, col, motion, True
            elif motion in [1,5] and v[row][col][0]==mine:# owner check + boundary check
                return row, col, motion, row>0
            elif motion in [2,6] and v[row][col][0]==mine:
                return row, col, motion, col<7
            elif motion in [3,7] and v[row][col][0]==mine:
                return row, col, motion, row<3
            elif motion in [4,8] and v[row][col][0]==mine:
                return row, col, motion, col>0
        raise
    except:
        return -1, -1, -1, False

def flip(row, col):
    global mine, flag
    v[row][col] = chess.pop()
    if not flag:
        mine = v[row][col][0]
        flag = True
        
def move(row, col, motion):
    if motion==1 and v[row-1][col]=="_":
        v[row][col], v[row-1][col] = v[row-1][col], v[row][col]
    elif motion==2 and v[row][col+1]=="_":
        v[row][col], v[row][col+1] = v[row][col+1], v[row][col]
    elif motion==3 and v[row+1][col]=="_":
        v[row][col], v[row+1][col] = v[row+1][col], v[row][col]
    elif motion==4 and v[row][col-1]=="_":
        v[row][col], v[row][col-1] = v[row][col-1], v[row][col]
    else:
        print("Not empty !")
        
def bigger(x, y):
    if x[0]==y[0]:
        return False
    if x[0]=='b':
        if x[1]=='將' and y[1]!='兵':
            return True
        elif x[1]=='士' and y[1]!='帥':
            return True
        elif x[1]=='象' and y[1] not in ['帥', '仕']:
            return True
        elif x[1]=='車' and y[1] in ['傌', '炮', '兵']:
            return True
        elif x[1]=='馬' and y[1] in ['炮', '兵']:
            return True
        elif x[1]=='包' and y[1]=='兵':
            return True
        elif x[1]=='卒' and y[1]=='帥':
            return True
    else:
        if x[1]=='帥' and y[1]!='卒':
            return True
        elif x[1]=='仕' and y[1]!='將':
            return True
        elif x[1]=='相' and y[1] not in ['將', '士']:
            return True
        elif x[1]=='俥' and y[1] in ['馬', '包', '卒']:
            return True
        elif x[1]=='傌' and y[1] in ['包', '卒']:
            return True
        elif x[1]=='炮' and y[1]=='卒':
            return True
        elif x[1]=='兵' and y[1]=='將':
            return True
    return False

def eat(row, col, motion):
    if motion==5 and bigger(v[row][col], v[row-1][col]):
        v[row-1][col] = v[row][col]
        v[row][col] = "_"
    elif motion==6 and bigger(v[row][col], v[row][col+1]):
        v[row][col+1] = v[row][col]
        v[row][col] = "_"
    elif motion==7 and bigger(v[row][col], v[row+1][col]):
        v[row+1][col] = v[row][col]
        v[row][col] = "_"
    elif motion==8 and bigger(v[row][col], v[row][col-1]):
        v[row][col-1] = v[row][col]
        v[row][col] = "_"
    else:
        print("Smaller !!")
```

```python
while True:
    printv()
    row = input("Enter row 0~3 : ")
    col = input("Enter col 0~7 : ")
    motion = input("Enter motion flip(0)/move(1~4)/eat(5~9) : ")
    row, col, motion, valid = check(row,col,motion)
    if valid:
        if motion==0:
            flip(row,col)
        elif motion in [1,2,3,4]:
            move(row,col,motion)
        else:
            eat(row,col,motion)
    else:
        print("Input error")
    print()
```
